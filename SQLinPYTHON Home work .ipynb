### Dumping the cardata into the table

```python
import csv
with open('car.data','r') as f:
    next(f)
    car_data=csv.reader(f,delimiter='\n')
    for rows in car_data:
        print(rows)
        rows_data=rows[0].split(",")
        print(rows[0])
        cursor.execute("""insert into tasktoprasanth.cardata3(buying,maint,doors,persons,lug_boot,safety,typeof) values(%s,%s,%s,%s,%s,%s,%s)""",rows_data)
```
